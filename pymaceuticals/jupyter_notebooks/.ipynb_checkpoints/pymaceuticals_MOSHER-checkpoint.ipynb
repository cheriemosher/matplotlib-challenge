```python
# Dependencies and setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# Load files
mouse_drug_data_to_load = "../data/mouse_drug_data.csv"
clinical_trial_data_to_load = "../data/clinicaltrial_data.csv"

# Read the Mouse and clinical trial data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
drug_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
drug_trials_data = pd.merge(mouse_data, drug_data, how = "left", 
                            on = ["Mouse ID", "Mouse ID"])

# Display the data table for preview
drug_trials_data
```

```python
# Set drug as index to reduce the size of the dataframe
drug_trials_data.set_index("Drug", inplace = True)

# Reduce dataframe to retain drugs for analysis
for_comparisons = drug_trials_data.loc[["Capomulin", "Infubinol", 
                                        "Ketapril", "Placebo"], :] 

# Remove 'Drug' as index for future data manipulation
for_comparisons.reset_index(inplace = True)
```

## Tumor Response to Treatment

```python
# Create a dataframe with 'Drug', 'Timepoint', and 'Tumor Volume' 
for_tumor = for_comparisons[["Drug", "Timepoint", "Tumor Volume (mm3)"]]

# Group by 'Drug' and 'Timepoint'
groupby_tumor_response = for_tumor.groupby(["Drug", "Timepoint"])
```

```python
# Calculate the mean tumor response by drug and per timepoint
mean_tumor_response = groupby_tumor_response.mean()

# Pivot the table so that the data reflects tumor volumes by drug, per timepoint 
move_tumor_response = pd.pivot_table(mean_tumor_response, 
                                     index = ["Timepoint"], columns = ["Drug"])

# Remove the 'Tumor Volume (mm3)' (makes it easier to graph drug with st. error later)
for_tumor_response = move_tumor_response["Tumor Volume (mm3)"]

# Reset index (makes it easier to graph drug with st. error later)
for_tumor_response = for_tumor_response.reset_index()

# Print head of adjusted dataframe
for_tumor_response.head()
```

```python
# Calculate the standard error of tumor response by drug and per timepoint
st_err_tumor_response = groupby_tumor_response.sem()

# Do the same process for the data based on standard error as for the data based on mean
move_err_tumor_response = pd.pivot_table(st_err_tumor_response, 
                                         index = ["Timepoint"], columns = ["Drug"])

for_err_tumor_response = move_err_tumor_response["Tumor Volume (mm3)"]

for_err_tumor_response = for_err_tumor_response.reset_index()

for_err_tumor_response.head()
```

```python
# Create a list with the x-axis values (the timepoints)
x_axis_tumor_response = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]
```

```python
# Create the plot area
fig, ax = plt.subplots()

# Plot the mean and standard error for each drug and the placebo
ax.errorbar(x_axis_tumor_response, for_tumor_response["Capomulin"], 
            for_err_tumor_response["Capomulin"], linewidth = 0.5, marker = "s", 
            label = "Capomulin")
ax.errorbar(x_axis_tumor_response, for_tumor_response["Infubinol"], 
            for_err_tumor_response["Infubinol"], linewidth = 0.5, marker = "v", 
            label = "Infubinol")
ax.errorbar(x_axis_tumor_response, for_tumor_response["Ketapril"], 
            for_err_tumor_response["Ketapril"], linewidth = 0.5, marker = "^", 
            label = "Ketapril")
ax.errorbar(x_axis_tumor_response, for_tumor_response["Placebo"], 
            for_err_tumor_response["Placebo"], linewidth = 0.5, marker = "o", 
            label = "Placebo")

# Set the x-axis range
ax.set_xlim(-5, 50)

# Label the x- and y-axes
ax.set_xlabel("Treatment Duration (Days)")
ax.set_ylabel("Tumor Volume (mm3)")

# Label the graph
ax.set_title("Tumor Response to Treatments")

# Place the legend in the best place on the plot and make the font readable
plt.legend(loc = "best", fontsize = "medium")

# Save the graph
plt.savefig("../images/tumor_response_treatment_MOSHER.png")
```

## Metastatic Response to Treatment

```python
# Create a dataframe with 'Drug', 'Timepoint', and 'Metastatic Sites'
for_metastatic = for_comparisons[["Drug", "Timepoint", "Metastatic Sites"]]

# Group by 'Drug' and 'Timepoint'
groupby_metastatic_response = for_metastatic.groupby(["Drug", "Timepoint"])
```

```python
# Create a dataframe of means ready for graphing, as was done in the previous section
mean_meta_response = groupby_metastatic_response.mean()

move_meta_response = pd.pivot_table(mean_meta_response, index = ["Timepoint"], 
                                    columns = ["Drug"])

for_meta_response = move_meta_response["Metastatic Sites"]

for_meta_response = for_meta_response.reset_index()

for_meta_response.head()
```

```python
# Create a dataframe of standard errors ready for graphing, as was done in the previous section
st_err_meta_response = groupby_metastatic_response.sem()

move_err_meta_response = pd.pivot_table(st_err_meta_response, index = ["Timepoint"], 
                                        columns = ["Drug"])

for_err_meta_response = move_err_meta_response["Metastatic Sites"]

for_err_meta_response = for_err_meta_response.reset_index()

for_err_meta_response.head()
```

```python
# x-axis, same as above
x_axis_meta_response = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]
```

```python
# Create a graph, same as the last section
fig, ax = plt.subplots()

ax.errorbar(x_axis_meta_response, for_meta_response["Capomulin"], 
            for_err_meta_response["Capomulin"], linewidth = 0.5, marker = "s", 
            label = "Capomulin")
ax.errorbar(x_axis_meta_response, for_meta_response["Infubinol"], 
            for_err_meta_response["Infubinol"], linewidth = 0.5, marker = "v", 
            label = "Infubinol")
ax.errorbar(x_axis_meta_response, for_meta_response["Ketapril"], 
            for_err_meta_response["Ketapril"], linewidth = 0.5, marker = "^", 
            label = "Ketapril")
ax.errorbar(x_axis_meta_response, for_meta_response["Placebo"], 
            for_err_meta_response["Placebo"], linewidth = 0.5, marker = "o", 
            label = "Placebo")

ax.set_xlim(-5, 50)

ax.set_xlabel("Treatment Duration (Days)")
ax.set_ylabel("Metastatic Sites")

ax.set_title("Metastatic Spread During Treatments")

plt.legend(loc = "best", fontsize = "medium")

plt.savefig("../images/metastatic_spread_MOSHER.png")
```

## Survival Rates

```python
# Create a dataframe with 'Drug', 'Timepoint', and 'Mouse ID'
for_survival = for_comparisons[["Drug", "Timepoint", "Mouse ID"]]

# Group by 'Drug' and 'Timepoint', and count mouse I.Ds per drug and timepoint
groupby_mouse_survival = for_survival.groupby(["Drug", "Timepoint"]).count()

# Per drug, calculate the percent of mice who survived at each timepoint
groupby_mouse_survival["Percent Survival"] = (groupby_mouse_survival["Mouse ID"] / 25) * 100 
```

```python
# Same as above, pivot table to prepare for graphing
for_mouse_survival = pd.pivot_table(groupby_mouse_survival, 
                                     index = ["Timepoint"], columns = ["Drug"])

for_mouse_survival = for_mouse_survival["Percent Survival"]

for_mouse_survival = for_mouse_survival.reset_index()

for_mouse_survival.head()
```

```python
# x-axis, same as above
x_axis_mouse_survival = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45]
```

```python
# Since there is no standard error for the graph, used 'plot'
plt.plot(x_axis_mouse_survival, for_mouse_survival["Capomulin"], marker = "s", 
         linewidth = 0.5, color = "blue", label = "Capomulin")
plt.plot(x_axis_mouse_survival, for_mouse_survival["Infubinol"], marker = "v", 
         linewidth = 0.5, color = "orange", label = "Infubinol")
plt.plot(x_axis_mouse_survival, for_mouse_survival["Ketapril"], marker = "^", 
         linewidth = 0.5, color = "green", label = "Ketapril")
plt.plot(x_axis_mouse_survival, for_mouse_survival["Placebo"], marker = "o", 
         linewidth = 0.5, color = "red", label = "Placebo")

plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Survival Rate (%)")

plt.title("Mouse Survival During Treatment")

plt.legend(loc = "best", fontsize = "medium")

plt.savefig("../images/survival_rates_MOSHER.png")
```

## Summary Bar Graph

```python
# Create a dataframe with 'Timepoint', the three drugs we are interested in, and the placebo
tumor_change = for_tumor_response[["Timepoint", "Capomulin", "Infubinol", 
                                     "Ketapril", "Placebo"]]

# Make the index: 'Timepoint' 
tumor_change.set_index("Timepoint", inplace = True)
```

```python
# Calculate the percent change from day 0 to day 45 for each drug (and the placebo)
percent_change = ((tumor_change.iloc[-1, :] - tumor_change.iloc[0, :]) 
                  / tumor_change.iloc[0, :]) * 100

# Round to two decimal places
percent_change = (round(percent_change, 2))

# Create a tuple of the data
percent_change_tuple = tuple(percent_change)
```

```python
# Create a list of x axis labels 
x_axis = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]

# Create an numbered list for the x axis
x_all = np.arange(len(percent_change_tuple))
```

```python
# Create the plot area 
fig, ax = plt.subplots()

# Plot the data as a bar plot
# Make the bar that is below zero green, and those above red
rects1 = ax.bar(x_all, percent_change_tuple, label = "x_axis", 
                color = ("g", "r", "r", "r"))

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel("% Change in Tumor Size")
ax.set_title("% Tumor Change Over 45 Day Treatment")
ax.set_xticks(x_all)
ax.set_xticklabels(x_axis)
ax.set_ylim(-30, 70)

# Add a line at y = zero
plt.hlines(0, -.5, 3.5)


# Add labels to the bars that represent their heights
# I don't yet understand this, but it worked!
def autolabel(rects, xpos = "center"):
    
    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0, 'right': 1, 'left': -1}

    for rect in rects:
        height = rect.get_height()
        ax.annotate("{}%".format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(offset[xpos]*3, 3),  # use 3 points offset
                    textcoords="offset points",  # in both directions
                    ha=ha[xpos], va='bottom')

autolabel(rects1, "center")

# Save figure
plt.savefig("../images/summary_bar_graph_MOSHER.png")
```
